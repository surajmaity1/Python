```python
sns.heatmap(df.corr())
```

```python
# JOINT PLOT

'''
In seaborn, jointplot() is a function that allows you to create 
a combination of univariate and bivariate plots to visualize the relationship between 
two variables in a dataset. 
It is a convenient way to explore the joint distribution of two variables,
their individual distributions, and their scatterplot all in one figure.
'''
```

```python
sns.jointplot(x='tip', y='total_bill', data = df, kind = 'hex')
```

```python
sns.jointplot(x='tip', y='total_bill', data=df, kind='reg')
```

```python
# PAIR PLOT
```

```python
'''
To plot multiple pairwise bivariate distributions in a dataset, you can use the .pairplot() function. 
In seaborn, pairplot() is a function that allows you to create a grid of scatterplots and histograms to visualize the pairwise relationships between multiple variables in a dataset. It provides a quick way to explore the relationships between variables and identify potential patterns or correlations.

'''
```

```python
sns.pairplot(df)
```

```python
sns.pairplot(df, hue = 'sex')
```

```python
sns.pairplot(df, hue = 'smoker')
```

```python
# DIST PLOT
```

```python
'''
distplot() is a function that allows you to create a histogram with a kernel density estimate (KDE) curve overlaid on top. It provides a convenient way to visualize the distribution of a single variable in a dataset, showing the shape and characteristics of its distribution.


KDE stands for Kernel Density Estimation and that is another kind of the plot in seaborn.

bins is used to set the number of bins you want in your plot and it actually depends on your dataset.
'''
```

```python
sns.distplot(df['tip'])
```

```python
sns.distplot(df['tip'],kde=True, bins=10)
```

```python
sns.distplot(df['tip'],kde=False, bins=10)
```
